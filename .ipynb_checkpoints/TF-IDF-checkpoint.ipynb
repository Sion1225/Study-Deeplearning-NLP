```python
import pandas as pd
from math import log
```

```python
docs = [
    "I want to eat apple",
    "I want to eat banana",
    "Yellow and long banana banana",
    "I like fruit"
]
vocab = list(set(w for doc in docs for w in doc.split()))
vocab.sort()
print(vocab)
```

```python
N = len(docs) #Number of document

def tf(t,d):
    return d.count(t)

def idf(t):
    df = 0
    for doc in docs:
        df += t in doc
    return log(N/(df+1))

def tfidf(t,d):
    return tf(t,d)*idf(t)
```

```python
result = []

for i in range(N):
    result.append([])
    d = docs[i]
    for j in range(len(vocab)):
        t = vocab[j]
        result[-1].append(tf(t,d))
        
tf_ = pd.DataFrame(result, columns = vocab)
```

```python
tf_
```

```python
result = []

for j in range(len(vocab)):
    t = vocab[j]
    result.append(idf(t))

idf_ = pd.DataFrame(result, index=vocab, columns=["IDF"])
idf_
```

```python
result = []
for i in range(N):
  result.append([])
  d = docs[i]
  for j in range(len(vocab)):
    t = vocab[j]
    result[-1].append(tfidf(t,d))

tfidf_ = pd.DataFrame(result, columns = vocab)
tfidf_
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
corpus  = [
    "I want to eat apple",
    "I want to eat banana",
    "Yellow and long banana banana",
    "I like fruit"
]

vector = CountVectorizer()
```

```python
#Chart Count of Voca.
print("DTM\n",vector.fit_transform(corpus).toarray(),"\n")

#Index
print(vector.vocabulary_,"\n")

print(vector.fit_transform(corpus))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
corpus  = [
    "I want to eat apple",
    "I want to eat banana",
    "Yellow and long banana banana",
    "I like fruit"
]

tfidf_v=TfidfVectorizer().fit(corpus)
print(tfidf_v.transform(corpus).toarray())
print(tfidf_v.vocabulary_)
```
