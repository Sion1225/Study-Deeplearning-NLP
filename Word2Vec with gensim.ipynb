```python
import re
import urllib.request
import zipfile
from lxml import etree
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
urllib.request.urlretrieve("https://raw.githubusercontent.com/ukairia777/tensorflow-nlp-tutorial/main/09.%20Word%20Embedding/dataset/ted_en-20160408.xml", filename="DataSet/ted_en-20160408.xml")
```

```python
targetXML = open('DataSet/ted_en-20160408.xml', 'r', encoding='UTF8')
target_text = etree.parse(targetXML)
```

```python
parse_text = '\n'.join(target_text.xpath('//content/text()'))
print("1",end=' ')

content_text = re.sub(r'\([^)]*\)', '', parse_text)
print("2",end=' ')

sent_text = sent_tokenize(content_text)
print("3",end=' ')

normalized_text = []
for string in sent_text:
    tokens = re.sub(r"[^a-z0-9]+", " ", string.lower())
    normalized_text.append(tokens)
    print(".",end=' ')
    
result = [word_tokenize(sentence) for sentence in normalized_text]
```

```python
print("Count of samples : ",len(result))
```

```python
for line in result[:5]:
    print(line)
```

```python
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
```

```python
model = Word2Vec(sentences=result, vector_size=100, window=5, min_count=5, workers=4, sg=0)
```

```python
model.wv.most_similar("man")
```

```python
model.wv.most_similar("companies")
```

```python
model.wv.save_word2vec_format('Models/eng_w2v')
```
